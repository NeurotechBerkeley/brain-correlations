```python
import imageio
import pylab
import numpy as np
%matplotlib inline
```

```python
fname = '../data/IMG_5745s.mov'
vid = imageio.get_reader(fname)
vid.get_length()
```

```python
sums = [vid.get_data(i).sum() for i in range(vid.get_length())]
sums = np.array(sums)
```

```python
img = vid.get_data(302)
fig = pylab.figure()
pylab.imshow(img)

img.sum()
```

```python
white_index = np.argmax(np.diff(sums/1000.0))+1
white_index
```
